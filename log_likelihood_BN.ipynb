```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

N = 20
vars = ["x", "y", "z"]

df = pd.DataFrame([np.round(np.random.random(3)) for _ in range(N)], 
    columns=vars).astype(int)

states = {
    "x": range(2),
    "y": range(2),
    "z": range(2)
}

res = {
    "x -> z -> y": 0,
    "x -> z <- y": 0,
    "x <- z -> y": 0,
    "x <- z <- y": 0,
}

df
```

```python

def log_likelihood(vars, pa):
    logl = 0
    for var in vars:
        for i in states[var]:
            if len(pa[var]) > 0:
                for parent in pa[var]:
                    for j in states[parent]:
                        logl += df[(df[var] == i) & (df[parent] == j)].shape[0] * np.log(df[(df[var] == i) & (df[parent] == j)].shape[0] / df[df[parent] == j].shape[0])

            else:
                logl += df[df[var] == i].shape[0] * np.log(df[df[var] == i].shape[0] / df.shape[0])
    return logl

res["x -> z -> y"] = log_likelihood(vars, {"x": [], "z": ["x"], "y": ["z"]})
res["x -> z <- y"] = log_likelihood(vars, {"x": [], "z": ["x", "y"], "y": []})
res["x <- z -> y"] = log_likelihood(vars, {"x": ["z"], "z": [], "y": ["z"]})
res["x <- z <- y"] = log_likelihood(vars, {"x": ["z"], "z": ["y"], "y": []})

res
```
